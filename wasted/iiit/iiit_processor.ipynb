```python
import pandas as pd
import re

# Dictionary mapping IIIT placement data with standardized JoSAA GFTI names
iiit_placement_dict = {
    # Standardized institute names as keys (matching JoSAA GFTI format)
    "Atal Bihari Vajpayee Indian Institute of Information Technology & Management Gwalior": {
        "average_package": 2056000,
        "highest_package": 6500000
    },
    "Indian Institute of Information Technology Allahabad": {
        "average_package": 2578000,
        "highest_package": 12100000
    },
    "Indian Institute of Information Technology Lucknow": {
        "average_package": 3371000,
        "highest_package": 14500000
    },
    "Indian Institute of Information Technology Design and Manufacturing Jabalpur": {
        "average_package": 1927000,
        "highest_package": 11000000
    },
    "Indian Institute of Information Technology Pune": {
        "average_package": 1360000,
        "highest_package": 4300000
    },
    "Indian Institute of Information Technology Tiruchirappalli": {
        "average_package": 1000000,
        "highest_package": 4600000
    },
    "Indian Institute of Information Technology Bhopal": {
        "average_package": 1677000,
        "highest_package": 8500000
    },
    "Indian Institute of Information Technology Sri City": {
        "average_package": 1844000,
        "highest_package": 12000000
    },
    "Indian Institute of Information Technology Ranchi": {
        "average_package": 1214000,
        "highest_package": 2800000
    },
    "Indian Institute of Information Technology Kottayam": {
        "average_package": 1191000,
        "highest_package": 5000000
    },
    "Indian Institute of Information Technology Bhagalpur": {
        "average_package": 957000,
        "highest_package": 8300000
    },
    "Indian Institute of Information Technology Design and Manufacturing Chennai": {
        "average_package": 1419000,
        "highest_package": 3164200
    },
    "Indian Institute of Information Technology Raichur": {
        "average_package": 1800000,
        "highest_package": 4500000
    },
    "Indian Institute of Information Technology(IIIT) Kalyani, West Bengal": {
        "average_package": 1072000,
        "highest_package": 4400000
    },
    "Indian Institute of Information Technology Surat": {
        "average_package": 1399000,
        "highest_package": 7400000
    },
    "Indian Institute of Information Technology Nagpur": {
        "average_package": 1311000,
        "highest_package": 4700000
    },
    "Indian Institute of Information Technology Una": {
        "average_package": 1102000,
        "highest_package": 3000000
    },
    "Indian Institute of Information Technology Vadodara": {
        "average_package": 1200000,
        "highest_package": 5630000
    },
    "Indian Institute of Information Technology Agartala": {
        "average_package": 1800000,
        "highest_package": 6000000
    },
    "Indian Institute of Information Technology(IIIT) Kilohrad, Sonepat, Haryana": {
        "average_package": 1587000,
        "highest_package": 4000000
    },
    "Indian Institute of Information Technology (IIIT)Kota, Rajasthan": {
        "average_package": 1309000,
        "highest_package": 6500000
    },
    "Indian Institute of Information Technology Guwahati": {
        "average_package": 1675000,
        "highest_package": 7100000
    },
    "Indian Institute of Information Technology Dharwad": {
        "average_package": 1190000,
        "highest_package": 7194000
    },
    "Indian Institute of Information Technology Design and Manufacturing Kurnool": {
        "average_package": 820000,
        "highest_package": 3250000
    },
    "Indian Institute of Information Technology Manipur (Imphal)": {
        "average_package": 870000,
        "highest_package": 4000000
    }
}

def normalize_institute_name(name):
    """
    Normalize institute names for better matching
    """
    if pd.isna(name):
        return ""
    
    # Convert to string and strip whitespace
    name = str(name).strip()
    
    # Remove extra spaces and normalize spacing
    name = re.sub(r'\s+', ' ', name)
    
    # Handle common variations
    name = name.replace('&', 'and')
    name = name.replace(' and Management', ' & Management')
    
    return name

def fuzzy_match_institute(institute_name, placement_dict):
    """
    Find the best match for an institute name in the placement dictionary
    using fuzzy matching based on city/location names
    """
    if not institute_name:
        return None
    
    institute_lower = institute_name.lower()
    
    # Extract city/location from institute name
    city_patterns = {
        'gwalior': ['gwalior'],
        'allahabad': ['allahabad'],
        'lucknow': ['lucknow'],
        'jabalpur': ['jabalpur'],
        'pune': ['pune'],
        'tiruchirappalli': ['tiruchirappalli', 'trichy'],
        'bhopal': ['bhopal'],
        'sri city': ['sri city'],
        'ranchi': ['ranchi'],
        'kottayam': ['kottayam'],
        'bhagalpur': ['bhagalpur'],
        'chennai': ['chennai'],
        'raichur': ['raichur'],
        'kalyani': ['kalyani'],
        'surat': ['surat'],
        'nagpur': ['nagpur'],
        'una': ['una'],
        'vadodara': ['vadodara'],
        'agartala': ['agartala'],
        'sonepat': ['sonepat', 'kilohrad'],
        'kota': ['kota'],
        'guwahati': ['guwahati'],
        'dharwad': ['dharwad'],
        'kurnool': ['kurnool'],
        'manipur': ['manipur', 'imphal']
    }
    
    # Try to match based on city/location
    for city, patterns in city_patterns.items():
        for pattern in patterns:
            if pattern in institute_lower:
                # Find matching entry in placement_dict
                for dict_key in placement_dict.keys():
                    if city in dict_key.lower() or any(p in dict_key.lower() for p in patterns):
                        return dict_key
    
    # Direct name matching as fallback
    normalized_input = normalize_institute_name(institute_name)
    for dict_key in placement_dict.keys():
        normalized_dict_key = normalize_institute_name(dict_key)
        if normalized_input == normalized_dict_key:
            return dict_key
    
    return None

def merge_placement_data(csv_file_path, output_file_path=None):
    """
    Merge placement data with IIIT data CSV
    
    Parameters:
    csv_file_path (str): Path to your IIIT data CSV file
    output_file_path (str): Path for output CSV file (optional)
    
    Returns:
    pandas.DataFrame: Merged dataframe
    """
    
    # Read the CSV file
    try:
        df = pd.read_csv(csv_file_path)
        print(f"Successfully loaded CSV with {len(df)} rows")
    except Exception as e:
        print(f"Error reading CSV file: {e}")
        return None
    
    # Check if 'institute' column exists
    if 'institute' not in df.columns:
        print("Error: 'institute' column not found in CSV")
        print(f"Available columns: {list(df.columns)}")
        return None
    
    # Initialize new columns
    df['average_package'] = None
    df['highest_package'] = None
    
    # Track matching statistics
    matched_institutes = set()
    unmatched_institutes = set()
    
    # Merge placement data
    for index, row in df.iterrows():
        institute_name = row['institute']
        
        # Find matching placement data
        matched_key = fuzzy_match_institute(institute_name, iiit_placement_dict)
        
        if matched_key:
            df.at[index, 'average_package'] = iiit_placement_dict[matched_key]['average_package']
            df.at[index, 'highest_package'] = iiit_placement_dict[matched_key]['highest_package']
            matched_institutes.add(institute_name)
        else:
            unmatched_institutes.add(institute_name)
    
    # Print matching statistics
    print(f"\nMatching Statistics:")
    print(f"Matched institutes: {len(matched_institutes)}")
    print(f"Unmatched institutes: {len(unmatched_institutes)}")
    
    if matched_institutes:
        print(f"\nMatched institutes:")
        for inst in sorted(matched_institutes):
            print(f"  - {inst}")
    
    if unmatched_institutes:
        print(f"\nUnmatched institutes:")
        for inst in sorted(unmatched_institutes):
            print(f"  - {inst}")
    
    # Save the merged data
    if output_file_path:
        try:
            df.to_csv(output_file_path, index=False)
            print(f"\nMerged data saved to: {output_file_path}")
        except Exception as e:
            print(f"Error saving CSV file: {e}")
    
    return df

# Example usage:
if __name__ == "__main__":
    # Replace 'your_iiit_data.csv' with your actual file path
    input_file = 'iiit_data.csv'
    output_file = 'iiit_data_with_placements.csv'
    
    # Merge the data
    merged_df = merge_placement_data(input_file, output_file)
    
    if merged_df is not None:
        # Display sample of merged data
        print(f"\nSample of merged data:")
        print(merged_df[['institute', 'program_name', 'average_package', 'highest_package']].head(10))
        
        # Show statistics
        total_rows = len(merged_df)
        rows_with_placement = len(merged_df[merged_df['average_package'].notna()])
        print(f"\nTotal rows: {total_rows}")
        print(f"Rows with placement data: {rows_with_placement}")
        print(f"Coverage: {(rows_with_placement/total_rows)*100:.1f}%")

# Alternative: If you want to use the dictionary directly in your existing code
def get_placement_data(institute_name):
    """
    Get placement data for a specific institute
    
    Parameters:
    institute_name (str): Name of the institute
    
    Returns:
    dict: Dictionary with average_package and highest_package, or None if not found
    """
    matched_key = fuzzy_match_institute(institute_name, iiit_placement_dict)
    if matched_key:
        return iiit_placement_dict[matched_key]
    return None

# Example of using the get_placement_data function:
# placement_info = get_placement_data("Indian Institute of Information Technology Guwahati")
# if placement_info:
#     print(f"Average Package: {placement_info['average_package']:,}")
#     print(f"Highest Package: {placement_info['highest_package']:,}")
```
