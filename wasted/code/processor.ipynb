```python
import pandas as pd

def clean_package(value):
    if isinstance(value, str) and '-' in value:
        parts = value.split('-')
        try:
            return (float(parts[0]) + float(parts[1])) / 2
        except:
            return None
    try:
        return float(value)
    except:
        return None
    


```
