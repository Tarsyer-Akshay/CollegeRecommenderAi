## Cleaning cutoff data

```python
import pandas as pd
import numpy as np
```

```python
cf = pd.read_csv("cutoff_data.csv")
```

```python
cf.head()
```

```python
cf["gender"].unique()
```

```python
cf['program_name'] = cf['academic_program'].str.extract(r'^(.*?)(?:\s*\(.*\))')
```

```python
cf["program_name"].unique()
```

```python
cf['duration'] = cf['academic_program'].str.extract(r'\((\d+ Years)')
```

```python
cf['duration'].unique()
```

```python
cf['degree_raw'] = cf['academic_program'].str.extract(r'\,\s*(.*?)\)$')
```

```python
cf['degree_raw'].unique()
```

```python
cf = cf.rename(columns={"college_id":"nirf_rank"})
```

```python
cf['program_name_2'] = cf['academic_program'].str.extract(r'^(.*?)(?:\s*\(\d+ Years, .*?\))')[0].str.strip()
```

```python
cf['opening_rank'] = pd.to_numeric(cf['opening_rank'], errors='coerce')  # invalid values become NaN

# Optional: You can also do the same for 'closing_rank' if needed
cf['closing_rank'] = pd.to_numeric(cf['closing_rank'], errors='coerce')
```

```python
cf.head()
```

```python
cf['program_name'] = cf['program_name'].str.replace(r'^(B\.?Tech in |BS in )', '', regex=True)
```

```python
cf.drop(columns=["academic_program","program_name_2"], inplace=True)
```

```python
print(cf.isnull().sum())
```

```python
cf.to_csv('cutoff_data.csv', index=False)
```
