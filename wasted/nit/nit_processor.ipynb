```python
import pandas as pd
```

```python
df1 = pd.read_csv("josaa_round1_nit_results.csv")
df2 = pd.read_csv("josaa_round2_nit_results.csv")
df3 = pd.read_csv("josaa_round3_nit_results.csv")
df4 = pd.read_csv("josaa_round4_nit_results.csv")
df5 = pd.read_csv("josaa_round5_nit_results.csv")
```

```python
df1['round_number'] = 1
df2['round_number'] = 2
df3['round_number'] = 3
df4['round_number'] = 4
df5['round_number'] = 5
```

```python
df1.head()
```

```python
df1 = df1.drop('Quota',axis=1)
df2 = df2.drop('Quota',axis=1)
df3 = df3.drop('Quota',axis=1)
df4 = df4.drop('Quota',axis=1)
df5 = df5.drop('Quota',axis=1)
```

```python
merged_df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
```

```python
merged_df["round_number"].unique()
```

```python
merged_df = merged_df.rename(columns={
    "Institute": "institute",
    "Academic Program": "academic_program",
    "Seat Type": "seat_type",
    "Gender": "gender",
    "Opening Rank": "opening_rank",
    "Closing Rank": "closing_rank"
})
```

```python
merged_df.head()
```

```python
merged_df['program_name_'] = merged_df['academic_program'].str.extract(r'^(.*?)(?:\s*\(\d+ Years, .*?\))')[0].str.strip()
```

```python
merged_df['duration'] = merged_df['academic_program'].str.extract(r'\((\d+ Years)')
```

```python
merged_df['degree_raw'] = merged_df['academic_program'].str.extract(r'\,\s*(.*?)\)$')
```

```python
merged_df['opening_rank'] = pd.to_numeric(merged_df['opening_rank'], errors='coerce')  # invalid values become NaN

# Optional: You can also do the same for 'closing_rank' if needed
merged_df['closing_rank'] = pd.to_numeric(merged_df['closing_rank'], errors='coerce')
```

```python
merged_df.drop(columns=['academic_program'],inplace=True)
```

```python
rf = pd.read_csv("rank_data_2.csv")
```

```python
rf.head()
```

```python
merged_df = merged_df.merge(rf[['institute', 'college_id']], on='institute', how='left')
```

```python
merged_df.head()
```

```python
merged_df.to_csv('nit_data.csv', index=False)
```
