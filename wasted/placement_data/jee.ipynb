```python
import pandas as pd

# Placement data (from the web search)
placement_data = {
    'institute': [
        'National Institute of Technology Agartala',
        'National Institute of Technology Hamirpur',
        'National Institute of Technology Karnataka, Surathkal',
        'National Institute of Technology Arunachal Pradesh',
        'National Institute of Technology, Jamshedpur',
        'National Institute of Technology, Srinagar',
        'National Institute of Technology, Tiruchirappalli',
        'National Institute of Technology, Uttarakhand',
        'Indian Institute of Engineering Science and Technology, Shibpur',
        'Indian Institute of Information Technology, Design & Manufacturing, Kancheepuram'
    ],
    'average_package': [
        5.0,    # NIT Agartala
        10.58,  # NIT Hamirpur
        16.25,  # NIT Surathkal (NITK)
        7.5,    # NIT Arunachal (median for MTech)
        12.63,  # NIT Jamshedpur
        13.8,   # NIT Srinagar
        27.2,   # NIT Trichy (CSE)
        8.03,   # NIT Uttarakhand
        None,   # IIEST Shibpur (data not available)
        None    # IIITDM Kancheepuram (data not available)
    ],
    'highest_package': [
        41,     # NIT Agartala
        205,    # NIT Hamirpur
        55,     # NIT Surathkal
        59,     # NIT Arunachal
        82,     # NIT Jamshedpur
        39,     # NIT Srinagar
        52.9,   # NIT Trichy
        12.5,   # NIT Uttarakhand
        None,   # IIEST Shibpur
        None    # IIITDM Kancheepuram
    ]
}

# Convert to DataFrame
placement_df = pd.DataFrame(placement_data)
```

```python
# Update only where values are missing
jee_mains.update(
    jee_mains.merge(
        placement_df,
        on='institute',
        how='left',
        suffixes=('', '_new')
    ).assign(
        average_package=lambda df: df['average_package'].combine_first(df['average_package_new']),
        highest_package=lambda df: df['highest_package'].combine_first(df['highest_package_new'])
    )[["average_package", "highest_package"]]
)
```

```python
import pandas as pd

placement_data = {
    'institute': [
        'National Institute of Technology, Andhra Pradesh',
    ],
    'average_package': [
        8.13,   # NIT Andhra Pradesh
        ],
    'highest_package': [
        44.1,   # NIT Andhra Pradesh
       
    ]
}

placement_df = pd.DataFrame(placement_data)

jee_updated = jee_mains.merge(
    placement_df,
    on='institute',
    how='left',
    suffixes=('', '_new')
).assign(
    average_package=lambda df: df['average_package'].combine_first(df['average_package_new']),
    highest_package=lambda df: df['highest_package'].combine_first(df['highest_package_new'])
)

# If you want to update in place:
jee_mains[['average_package', 'highest_package']] = jee_updated[['average_package', 'highest_package']]
```
